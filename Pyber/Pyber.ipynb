```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set csv paths
city_path = 'raw_data/city_data.csv'
ride_path = 'raw_data/ride_data.csv'
```

```python
# Read city csv
city_df = pd.read_csv(city_path)
city_df.head()
```

```python
city_df = city_df.sort_values(by='city')
city_df
```

```python
# Check column counts
city_df.count()
```

```python
# Read ride csv
ride_df = pd.read_csv(ride_path)
ride_df.head()
```

```python
# Check column counts
ride_df.count()
```

```python
ride_df = ride_df.sort_values(by='city')
ride_df['city']
```

```python
# Merge on city
pyber_df = pd.merge(ride_df, city_df, on='city', how='inner')
pyber_df.count()
```
